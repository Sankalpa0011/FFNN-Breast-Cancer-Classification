## Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
tf.random.set_seed(3)
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
```

## Data Coleection And Preprocessing

```python
cancer_data = sklearn.datasets.load_breast_cancer()
print(cancer_data)
```

```python
# Load data to a dataframe
cancer_dataset = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
cancer_dataset.head()
```

```python
# Adding the target column to the dataframe
cancer_dataset["label"] = cancer_data.target
cancer_dataset.tail()
```

```python
# Shape of the dataset
cancer_dataset.shape
```

```python
# Basic information about the dataset
cancer_dataset.info()
```

```python
# Check missing values
cancer_dataset.isna().sum()
```

```python
# Statistical measures
cancer_dataset.describe()
```

```python
cancer_dataset["label"].value_counts()
```

#### 0 --> Malignant

```python
cancer_dataset.groupby("label").mean()
```

## Target Feature Splitting

```python
x = cancer_dataset.drop(columns=["label"], axis=1)
y = cancer_dataset["label"]
```

## Train Test Splitting

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)
```

## Standardize The Data

```python
scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)

x_train = x_train_std
x_test = x_test_std
```

## Neaural Network Architecture

```python
model = Sequential()

model.add(Flatten(input_shape=(30,)))
model.add(Dense(64, activation="relu"))
model.add(Dense(2, activation="sigmoid"))

# Compile
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

# Summary
model.summary()
```

```python
# Train the neural network
history = model.fit(x_train, y_train, validation_split=0.1, epochs=30, batch_size=1)
```

## Model Evaluation

```python
# Visualizing the accuracy
plt.plot(history.history["accuracy"], label="Accuracy")
plt.plot(history.history["val_accuracy"], label="Val_Accuracy")
plt.title("Model Accuracy")
plt.xlabel("accuracy")
plt.ylabel("epoch")
plt.legend()
plt.show()
```

```python
# Visualizing the loss
plt.plot(history.history["loss"], label="Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.title("Model Loss")
plt.xlabel("loss")
plt.ylabel("epoch")
plt.legend()
plt.show()
```

## Accuracy Of The Model On Test Data

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(loss)
print(accuracy)
```
